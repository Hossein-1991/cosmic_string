````python
import numpy as np
import cv2 
from scipy.ndimage import gaussian_filter as gf
import torch
from torch.utils.data import TensorDataset
````

In this stage, we want to create our data! Without the doubt, the best choice is ```np.concatenate```. 





````python
for i in range(20):
  data = np.load('/content/drive/MyDrive/CMB_string/CMB_no_string_{}.npy'.format(i))
  img = cv2.flip(data, 0)
  img_2 = cv2.flip(data,1)
  img_3 = cv2.flip(img_2,0)
  img_4 = gf(data, sigma= 4)
  data = np.expand_dims(data, axis = 0) # For the feature extraction, we must add an additional dimesnsion here
  img = np.expand_dims(img, axis = 0)
  img_2 = np.expand_dims(img_2, axis = 0)
  img_3 = np.expand_dims(img_3, axis = 0)
  img_4 = np.expand_dims(img_4, axis = 0)
  if i == 0:
    dataa = np.concatenate((data,img,img_2,img_3,img_4), axis = 0)
  else:
    dataa = np.concatenate((dataa,data,img,img_2,img_3,img_4), axis = 0)
print (dataa.shape)
np.save('/content/drive/MyDrive/CMB_string/no_string',dataa)

````

Since data with string has different name, I am forced to write another cell for it!:

````python
c = 1
for i in range (5,10): # our data ranges from 5 to 9 (Gmu)
  for j in range (20): # each class has 20 members
    data = np.load('/content/drive/MyDrive/CMB_string/CMB_string_1e-0{}_gmu_{}.npy'.format(i,j))
    img = cv2.flip(data, 0)
    img_2 = cv2.flip(data,1)
    img_3 = cv2.flip(img_2,0)
    img_4 = gf(data, sigma= 4)
    data = np.expand_dims(data, axis = 0)
    img = np.expand_dims(img, axis = 0)
    img_2 = np.expand_dims(img_2, axis = 0)
    img_3 = np.expand_dims(img_3, axis = 0)
    img_4 = np.expand_dims(img_4, axis = 0)
    if j == 0:
      ddata = np.concatenate((data,img,img_2,img_3,img_4), axis = 0)
    else:
      ddata = np.concatenate((ddata,data,img,img_2,img_3,img_4), axis = 0)
  print (ddata.shape)
  np.save('/content/drive/MyDrive/CMB_string/string{}'.format(i),ddata)
  if c == 1:
    dd = np.concatenate((dataa,ddata),axis = 0)
    c = 2
  else:
    dd = np.concatenate((dd,ddata),axis = 0)
print (dd.shape) # dd is our final data!
````

````python
# creating labels:
lbl = np.full(100,0)
lbl5 = np.full(100,1)
lbl6 = np.full(100,2)
lbl7 = np.full(100,3)
lbl8 = np.full(100,4)
lbl9 = np.full(100,5)
label = np.concatenate((lbl,lbl5,lbl6,lbl7,lbl8,lbl9), axis = 0)
print (label.shape)
````

````python
dataset = TensorDataset(torch.tensor(dd).unsqueeze(1), torch.tensor(label))
torch.save(dataset, '/content/drive/MyDrive/CMB_string/dataset')
````
