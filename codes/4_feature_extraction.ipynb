```python
dataset = torch.load('/content/drive/MyDrive/CMB_string/dataset')
```

```python
class ff(nn.Module):
    def __init__(self):
        super(ff, self).__init__()

        # bring googlenet
        self.model = models.resnet18(pretrained= True) # recalling the resnet18 network
        self.model.conv1 = nn.Conv2d(1, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)
        self.model.fc = nn.Linear(512 , 50, bias = True)
        
    def forward(self, x):
        out = self.model(x)
        return out
```

```python
feature = ff()
```

```python
train_loader = DataLoader(dataset, batch_size=64, shuffle = 1)
x,y = next(iter(train_loader))
print (x.shape)
print (y.shape)
```

```python
for j, (image, lbl) in enumerate(train_loader):
  out = feature(image)
  torch.save(out,'/content/drive/MyDrive/out/{}'.format(j))
  torch.save(lbl,'/content/drive/MyDrive/out/L{}'.format(j))
```
