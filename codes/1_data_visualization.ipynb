```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
import seaborn as sns
cmap = plt.cm.get_cmap('jet')
```

Loading data from Mydrive:

```python
n_str = np.load('/content/drive/MyDrive/CMB_string/CMB_no_string_0.npy')
y_str = np.load('/content/drive/MyDrive/CMB_string/CMB_string_1e-05_gmu_0.npy')
yy_str = np.load('/content/drive/MyDrive/CMB_string/CMB_string_1e-06_gmu_0.npy')
```

```python
print (n_str.shape)
```

Lets depict these arrays (note that there are 6 different classes, however, I illustrate just three of them):

```python
plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
plt.title('no string')
plt.imshow(n_str,cmap = cmap)
plt.subplot(1,3,2)
plt.title('string with Gmu = 1e-05 ')
plt.imshow(y_str, cmap = cmap)
plt.subplot(1,3,3)
plt.title('string with Gmu = 1e-06 ')
plt.imshow(yy_str, cmap = cmap)
```

It may be useful to look at data distribution:

```python
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
plt.title('no string')
sns.histplot(n_str.flatten(), stat = 'density', kde = 1)
plt.subplot(1,3,2)
plt.title('string with Gmu = 1e-05 ')
sns.histplot(y_str.flatten(), stat = 'density', kde = 1)
plt.subplot(1,3,3)
plt.title('string with Gmu = 1e-06 ')
sns.histplot(yy_str.flatten(), stat = 'density', kde = 1)
```

As you can see, all three classes follow a Gaussian distribution with approximately the same mean (but different variance)
