```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy.ndimage import gaussian_filter as gf
cmap = plt.cm.get_cmap('jet')
```

```python
n_str = np.load('/content/drive/MyDrive/CMB_string/CMB_no_string_0.npy') 
```

When your data is limited and you want to prevent overfitting, the best choice is data augmentation in which you rotate your image in different direction. In this way, you generate new data which will help your system:

```python
# horizontal flip
img = cv2.flip(n_str, 0)
# vertical flip
img_2 = cv2.flip(n_str,1)
# combined flip
img_3 = cv2.flip(img_2,0)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title('no flip')
plt.imshow(n_str,cmap = cmap)
plt.subplot(2,2,2)
plt.title('horizontal flip ')
plt.imshow(img, cmap = cmap)
plt.subplot(2,2,3)
plt.title('vertical flip')
plt.imshow(img_2, cmap = cmap)
plt.subplot(2,2,4)
plt.title('combined flip')
plt.imshow(img_3, cmap = cmap)
```

Another useful approach is Gaussian blurring. In this way, you remove details of the image and highlite the most important aspects:

```python
img_4 = gf(n_str, sigma= 4)
plt.imshow(img_4, cmap = cmap)
```
